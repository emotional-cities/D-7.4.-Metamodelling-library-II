# Visualizations

Notebook employed to visualize the results of the metamodel by highlighting the regions where the output is higher.

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

with open('Data/X10_4.pickle', 'rb') as file:
    X = pickle.load(file)

with open('Data/Xtest20_4.pickle', 'rb') as file:
    X_test = pickle.load(file)

with open('Y_train_1.pickle', 'rb') as file:
    y1 = pickle.load(file)


with open('Y_train_2.pickle', 'rb') as file:
    y2 = pickle.load(file)

with open('Prediction_test_2_outputs.pickle', 'rb') as file:
    predictions = pickle.load(file)

percentile = np.percentile(predictions, 80, axis=0)
```

```python
Y_mask= predictions[:,0] > percentile[0]
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

axes[0].scatter(X_test[~Y_mask,0], X_test[~Y_mask,2], c='blue', alpha =0.6)
axes[0].scatter(X_test[Y_mask,0], X_test[Y_mask,2], c='red', label= 'Vulnerable cases', alpha =0.6)
axes[0].set_xlabel('beta_female_other')
axes[0].set_ylabel('beta_female_travel')
axes[0].legend()
axes[1].scatter(X_test[~Y_mask,0], X_test[~Y_mask,1], c='blue', alpha =0.6)
axes[1].scatter(X_test[Y_mask,0], X_test[Y_mask,1], c='red', label= 'Vulnerable cases', alpha =0.6)
axes[1].set_xlabel('beta_female_other')
axes[1].set_ylabel('beta_female_shop')
axes[1].legend()
axes[2].scatter(X_test[~Y_mask,1], X_test[~Y_mask,2], c='blue', alpha =0.6)
axes[2].scatter(X_test[Y_mask,1], X_test[Y_mask,2], c='red', label= 'Vulnerable cases', alpha =0.6)
axes[2].set_xlabel('beta_female_shop')
axes[2].set_ylabel('beta_female_travel')
axes[2].legend()
axes[3].scatter(X_test[~Y_mask,1], X_test[~Y_mask,3], c='blue', alpha =0.6)
axes[3].scatter(X_test[Y_mask,1], X_test[Y_mask,3], c='red', label= 'Vulnerable cases', alpha =0.6)
axes[3].set_xlabel('beta_female_shop')
axes[3].set_ylabel('female_income')
axes[3].legend()
plt.tight_layout()
fig.suptitle('Predicted difference in percentage of shopping trips between men and women', fontsize=12, fontweight='bold', y=1)
```
